```python
# import statements
import numpy as np
import matplotlib.pyplot as plt
```

# Part One

```python
def find_absorp_coef(strength, alpha0, v, v0, p, p0):
    """
    This function finds the absorption coefficient based on wavenumber.
    It assumes that the atmosphere is isothermal, thus the broadening
    of the line is only dependent on pressure. 
    
    strength = strength of the absorption line
    alpha0 = half width of the line
    v = wavenumber where you are measuring
    v0 = wavenumber of absorption line
    p = pressure
    p0 = 1000 hPa
    """

    alpha = alpha0*(p/p0)
    absorb_coef = (strength/np.pi)*(alpha/((v-v0)**2 + alpha**2))
    return(absorb_coef)

# find absorption coefficient from first line
strength = 1.132e-2 # atm/cm^2
p = 500 # pressure stated in question hPa
p0 = 1000 # surface pressure hPa
alpha0 = 0.068 # cm^-1
v = 7283.38 # cm^-1
v0 = 7283.02 # cm^-1
k1 = find_absorp_coef(strength, alpha0, v, v0, p, p0)
print('Absorption coefficient from first band: ', k1)


# now find absorption coefficient from second line
strength = 1.71e-2 # atm/cm^2
p = 500 # pressure stated in question hPa
p0 = 1000 # surface pressure hPa
alpha0 = 0.071 # cm^-1
v = 7283.38 # cm^-1
v0 = 7283.73 # cm^-1
k2 = find_absorp_coef(strength, alpha0, v, v0, p, p0)
print('Absorption coefficient from second band: ', k2)

# show the sum of both lines
print('Absorption coefficient from both bands: ', k1 + k2)
```

# Part Two

```python
def find_absorp_coef(strength, alpha0, v, v0, p, p0):
    alpha = alpha0*(p/p0)
    absorp_coef = (strength/np.pi)*(alpha/((v-v0)**2 + alpha**2))
    return(absorp_coef)

# set up evenly spaced wave numbers
wave_numbers = np.arange(start=7282, stop=7285, step=0.01)

# iterate over all wave numbers and find the sum as was done in part 1
absorp_coefs_by_v = []
for vn in wave_numbers:
    # find absorption coefficient from first line
    strength = 1.132e-2 # atm/cm^2
    p = 1000
    p0 = 1000
    alpha0 = 0.068 # cm^-1
    v = vn
    v0 = 7283.02 # cm^-1
    k1 = find_absorp_coef(strength, alpha0, v, v0, p, p0)


    # now find absorption coefficient from second line
    strength = 1.71e-2 # atm/cm^2
    p = 1000
    p0 = 1000
    alpha0 = 0.071 # cm^-1
    v = vn
    v0 = 7283.73 # cm^-1
    k2 = find_absorp_coef(strength, alpha0, v, v0, p, p0)

    # show the sum of both lines
    absorp_coef = k1 + k2

    # append this answer to a list of saved absorption coefficients
    absorp_coefs_by_v.append(absorp_coef)
```

# Part Three

```python

plt.plot(wave_numbers, absorp_coefs_by_v, color='black')
plt.xlabel('Wavenumbers (1/cm)')
plt.ylabel('Absorption Coefficient')
plt.ylim(0,0.1)
plt.grid()
plt.title('Absorption Coefficient as a Function of Wavenumber')
```
